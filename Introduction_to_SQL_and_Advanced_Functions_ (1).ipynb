1.  Explain the fundamental differences between DDL, DML, and DQL
commands in SQL. Provide one example for each type of command.



SQL commands are categorized into three main types based on their functionality:

### 1. Data Definition Language (DDL)

*   **Purpose:** DDL commands are used to define, modify, or delete database structures. They deal with the schema of the database rather than the data itself.
*   **Impact:** Changes the structure of tables, databases, views, etc.
*   **Examples of DDL commands:** `CREATE`, `ALTER`, `DROP`, `TRUNCATE`, `RENAME`.
*   **Example:**
    ```sql
    CREATE TABLE Students (
        StudentID INT PRIMARY KEY,
        FirstName VARCHAR(50),
        LastName VARCHAR(50),
        Age INT
    );
    ```
    (This command creates a new table named `Students` with specified columns and data types.)

### 2. Data Manipulation Language (DML)

*   **Purpose:** DML commands are used to manage data within the database objects. They deal with the actual data stored in the tables.
*   **Impact:** Inserts, updates, or deletes records within tables.
*   **Examples of DML commands:** `INSERT`, `UPDATE`, `DELETE`.
*   **Example:**
    ```sql
    INSERT INTO Students (StudentID, FirstName, LastName, Age)
    VALUES (1, 'Alice', 'Smith', 20);
    ```
    (This command inserts a new row of data into the `Students` table.)

### 3. Data Query Language (DQL)

*   **Purpose:** DQL commands are used to retrieve data from the database. It is essentially for querying and fetching information.
*   **Impact:** Retrieves data from one or more tables based on specified criteria.
*   **Examples of DQL commands:** `SELECT`.
*   **Example:**
    ```sql
    SELECT FirstName, LastName
    FROM Students
    WHERE Age > 19;
    ```
    (This command retrieves the `FirstName` and `LastName` of all students whose `Age` is greater than 19.)

### Purpose of SQL Constraints

SQL constraints are rules enforced on data columns in a table. They are used to limit the type of data that can go into a table, ensuring the accuracy and reliability of the data. Essentially, constraints maintain the data integrity of the database.

### Three Common Types of Constraints:

1.  **PRIMARY KEY**
    *   **Description:** A PRIMARY KEY constraint uniquely identifies each record in a table. Primary keys must contain unique values, and they cannot contain NULL values. Each table can have only one primary key, which can consist of single or multiple columns.
    *   **Scenario:** In a `Customers` table, a `CustomerID` column would be an excellent candidate for a PRIMARY KEY. This ensures that every customer has a unique identifier, and there are no duplicate `CustomerID` entries, preventing confusion and maintaining distinct customer records.

2.  **FOREIGN KEY**
    *   **Description:** A FOREIGN KEY is a field (or collection of fields) in one table that refers to the PRIMARY KEY in another table. It establishes and enforces a link between two tables, ensuring referential integrity.
    *   **Scenario:** Consider an `Orders` table and a `Customers` table. The `Orders` table could have a `CustomerID` column defined as a FOREIGN KEY referencing the `CustomerID` (PRIMARY KEY) in the `Customers` table. This prevents an order from being placed for a non-existent customer, ensuring that every order is associated with a valid customer.

3.  **NOT NULL**
    *   **Description:** The NOT NULL constraint ensures that a column cannot have a NULL value. This means that a value must always be provided for that column when a new record is inserted or an existing record is updated.
    *   **Scenario:** In a `Products` table, a `ProductName` column should ideally be defined as NOT NULL. This ensures that every product entered into the database has a name, preventing records with missing product names which could lead to data ambiguity or operational issues.

### Difference between `LIMIT` and `OFFSET` in SQL

Both `LIMIT` and `OFFSET` clauses are used with the `SELECT` statement to control which rows are returned from a query, but they serve different purposes:

*   **`LIMIT` Clause:**
    *   **Purpose:** `LIMIT` is used to restrict the number of rows returned by a query. It specifies the maximum number of rows that the `SELECT` statement should return.
    *   **Syntax Example:** `SELECT * FROM table_name LIMIT 10;` (This would return the first 10 rows).

*   **`OFFSET` Clause:**
    *   **Purpose:** `OFFSET` is used to skip a specified number of rows from the beginning of the result set before returning the remaining rows. It's often used in conjunction with `LIMIT` for pagination.
    *   **Syntax Example:** `SELECT * FROM table_name OFFSET 20;` (This would skip the first 20 rows and return all subsequent rows).

### Using `LIMIT` and `OFFSET` for Pagination (Retrieving the Third Page)

To retrieve a specific page of results, you combine `LIMIT` and `OFFSET`. The logic is:

*   `LIMIT` specifies the size of each page (number of records per page).
*   `OFFSET` specifies how many records to skip before starting to fetch the current page. The formula for `OFFSET` for an `N`th page with `P` records per page is `(N - 1) * P`.

Given that each page has 10 records and you want to retrieve the **third page**:

1.  **Page size (LIMIT):** 10 records.
2.  **Page number (N):** 3.
3.  **Records to skip (OFFSET):** `(3 - 1) * 10 = 2 * 10 = 20` records.

So, the SQL query to retrieve the third page of results would be:

```sql
SELECT *
FROM your_table_name
ORDER BY some_column -- It's crucial to have an ORDER BY for consistent pagination
LIMIT 10 OFFSET 20;
```

*   `LIMIT 10`: Retrieves 10 records.
*   `OFFSET 20`: Skips the first 20 records (which constitute the first two pages).

This query will effectively fetch records 21 through 30, which form the third page of results.

benefits? Provide a simple SQL example demonstrating its usage.

### What is a Common Table Expression (CTE) in SQL?

A Common Table Expression (CTE) is a temporary, named result set that you can reference within a single `SELECT`, `INSERT`, `UPDATE`, or `DELETE` statement. It's essentially a temporary view that exists only for the duration of a single query. CTEs are defined using the `WITH` clause.

### Main Benefits of CTEs:

1.  **Readability and Organization:** CTEs break down complex queries into logical, more manageable, and readable steps. Instead of writing one very long and nested query, you can define intermediate result sets, making the overall query easier to understand and debug.
2.  **Reusability within a Single Query:** A CTE can be referenced multiple times within the same query, which can reduce redundancy and make the query more efficient, especially if the sub-query would otherwise be repeated.
3.  **Simplifying Complex Joins and Subqueries:** They can simplify complex joins and make recursive queries possible (recursive CTEs are a powerful feature for hierarchical data).
4.  **Improved Performance (sometimes):** While not a guaranteed performance enhancer, using CTEs can sometimes allow the query optimizer to process the query more efficiently, especially by avoiding repeated calculations of the same sub-query.

### Simple SQL Example Demonstrating CTE Usage:

Let's say we have a table called `Employees` and we want to find the average salary of employees in each department, and then identify employees who earn more than their department's average salary.

```sql
-- Sample Data (for context, not part of the CTE)
CREATE TABLE Employees (
    EmployeeID INT PRIMARY KEY,
    Name VARCHAR(50),
    Department VARCHAR(50),
    Salary DECIMAL(10, 2)
);

INSERT INTO Employees (EmployeeID, Name, Department, Salary) VALUES
(1, 'Alice', 'HR', 60000.00),
(2, 'Bob', 'HR', 75000.00),
(3, 'Charlie', 'Sales', 80000.00),
(4, 'David', 'Sales', 65000.00),
(5, 'Eve', 'Marketing', 70000.00),
(6, 'Frank', 'Marketing', 72000.00),
(7, 'Grace', 'HR', 68000.00);

-- CTE Example:
WITH DepartmentAvgSalary AS (
    SELECT
        Department,
        AVG(Salary) AS AvgDeptSalary
    FROM
        Employees
    GROUP BY
        Department
)
SELECT
    e.Name,
    e.Department,
    e.Salary,
    das.AvgDeptSalary
FROM
    Employees e
JOIN
    DepartmentAvgSalary das ON e.Department = das.Department
WHERE
    e.Salary > das.AvgDeptSalary;
```

**Explanation of the example:**

1.  **`WITH DepartmentAvgSalary AS (...)`**: This defines a CTE named `DepartmentAvgSalary`.
2.  **Inside the CTE**: We calculate the average salary for each department by grouping the `Employees` table by `Department`.
3.  **Main Query**: We then join the `Employees` table (`e`) with our newly defined CTE `DepartmentAvgSalary` (`das`) on the `Department` column.
4.  **Filtering**: Finally, we filter the results to only include employees whose `Salary` is greater than their `AvgDeptSalary`.

This approach is much cleaner than writing a subquery in the `FROM` clause or repeating the average salary calculation for each department.

### SQL Normalization: Concept and Primary Goals

**Normalization** is a database design technique used to organize tables in a relational database to minimize data redundancy and improve data integrity. It involves breaking down a large table into smaller, more manageable tables and defining relationships between them.

**Primary Goals of Normalization:**

1.  **Reduce Data Redundancy:** By eliminating duplicate data, normalization saves storage space and prevents inconsistencies that can arise when the same data is stored in multiple places and only some instances are updated.
2.  **Improve Data Integrity:** It ensures that data dependencies are logical and that data is consistent across the database. This means that changes made to the data are less likely to cause errors or lead to incorrect information.
3.  **Eliminate Anomalies:**
    *   **Insertion Anomaly:** Difficulty in inserting new data if it depends on other data that is not yet available.
    *   **Deletion Anomaly:** Loss of critical data when a record is deleted because it was the only record containing that specific information.
    *   **Update Anomaly:** Inconsistencies that occur when updating redundant data, where not all copies of the data are updated.
4.  **Enhance Query Efficiency (indirectly):** While more joins might be required, smaller, more focused tables can sometimes lead to faster query execution, especially for update and insert operations.

### The First Three Normal Forms (1NF, 2NF, 3NF)

Normalization is typically applied in stages, with each stage representing a 'normal form'. The most common ones are 1NF, 2NF, and 3NF.

1.  **First Normal Form (1NF)**
    *   **Definition:** A table is in 1NF if:
        *   Each column contains atomic (indivisible) values. There are no repeating groups or multi-valued attributes.
        *   Each column has a unique name.
        *   The order of data does not matter.
    *   **Explanation:** This means that each cell in the table should contain a single value, not a list of values or a complex structure. For example, a `PhoneNumbers` column should not contain `"123-456-7890, 987-654-3210"`; instead, each phone number would be in its own row or a separate table.
    *   **Example (Not 1NF to 1NF):**
        | OrderID | Product        |
        |---------|----------------|
        | 1       | A, B, C        |
        | 2       | D, E           |

        _Becomes:_

        | OrderID | Product |
        |---------|---------|
        | 1       | A       |
        | 1       | B       |
        | 1       | C       |
        | 2       | D       |
        | 2       | E       |

2.  **Second Normal Form (2NF)**
    *   **Definition:** A table is in 2NF if:
        *   It is in 1NF.
        *   All non-key attributes are fully functionally dependent on the entire primary key. This means no non-key attribute is dependent on only a part of a composite primary key.
    *   **Explanation:** If a table has a composite primary key (a key made of two or more columns), 2NF ensures that any column that is not part of the primary key must depend on *all* parts of that composite key. If a non-key attribute only depends on a subset of the composite key, it should be moved to a separate table.
    *   **Example (Not 2NF to 2NF):**
        _Assume (OrderID, ProductID) is the composite primary key_

        | OrderID | ProductID | Product_Name | Customer_Name |
        |---------|-----------|--------------|---------------|
        | 1       | 101       | Laptop       | Alice         |
        | 1       | 102       | Mouse        | Alice         |

        Here, `Product_Name` depends only on `ProductID`, not on the full `(OrderID, ProductID)` composite key.

        _Becomes:_

        **Orders Table**
        | OrderID | Customer_Name |
        |---------|---------------|
        | 1       | Alice         |

        **Order_Details Table**
        | OrderID | ProductID |
        |---------|-----------|
        | 1       | 101       |
        | 1       | 102       |

        **Products Table**
        | ProductID | Product_Name |
        |-----------|--------------|
        | 101       | Laptop       |
        | 102       | Mouse        |

3.  **Third Normal Form (3NF)**
    *   **Definition:** A table is in 3NF if:
        *   It is in 2NF.
        *   There are no transitive dependencies. That is, no non-key attribute is dependent on another non-key attribute.
    *   **Explanation:** 3NF eliminates columns that are not directly dependent on the primary key but are dependent on another non-key column. If a non-key attribute can be determined by another non-key attribute, then those attributes should be moved to a separate table.
    *   **Example (Not 3NF to 3NF):**
        _Assume `EmployeeID` is the primary key_

        | EmployeeID | Employee_Name | Department | Dept_Head |
        |------------|---------------|------------|-----------|
        | 1          | John Doe      | Sales      | Jane Smith|
        | 2          | Mary Brown    | Sales      | Jane Smith|

        Here, `Dept_Head` is dependent on `Department`, and `Department` is a non-key attribute.

        _Becomes:_

        **Employees Table**
        | EmployeeID | Employee_Name | Department |
        |------------|---------------|------------|
        | 1          | John Doe      | Sales      |
        | 2          | Mary Brown    | Sales      |

        **Departments Table**
        | Department | Dept_Head |
        |------------|-----------|
        | Sales      | Jane Smith|


7.Generate a report showing CustomerName, Email, and the
TotalNumberofOrders for each customer. Include customers who have not placed
any orders, in which case their TotalNumberofOrders should be 0. Order the results
by CustomerName.

````python
print("\n--- Customer Order Report ---")

cursor.execute('''
SELECT
    c.CustomerName,
    c.Email,
    IFNULL(COUNT(o.OrderID), 0) AS TotalNumberofOrders
FROM
    Customers c
LEFT JOIN
    Orders o ON c.CustomerID = o.CustomerID
GROUP BY
    c.CustomerID, c.CustomerName, c.Email
ORDER BY
    c.CustomerName;
''')

report_results = cursor.fetchall()

# Print header
print(f"{"Customer Name":<20} {"Email":<25} {"Total Orders":<15}")
print("-" * 60)

# Print results
for row in report_results:
    print(f"{row[0]:<20} {row[1]:<25} {row[2]:<15}")
````

### Explanation of the SQL Query:

*   **`SELECT c.CustomerName, c.Email, IFNULL(COUNT(o.OrderID), 0) AS TotalNumberofOrders`**:
    *   We select the `CustomerName` and `Email` directly from the `Customers` table.
    *   `COUNT(o.OrderID)` counts the number of orders for each customer.
    *   `IFNULL(..., 0)` is crucial here: if a customer has no matching orders (due to the `LEFT JOIN`), `COUNT(o.OrderID)` would return `NULL`. `IFNULL` replaces these `NULL`s with `0`, satisfying the requirement to show 0 orders for such customers.
    *   `AS TotalNumberofOrders` assigns a friendly alias to the calculated count.

*   **`FROM Customers c LEFT JOIN Orders o ON c.CustomerID = o.CustomerID`**:
    *   We start by selecting from the `Customers` table (aliased as `c`).
    *   `LEFT JOIN Orders o` ensures that **all** customers from the `Customers` table are included in the result set, even if they have no corresponding entries in the `Orders` table. If there's no match, the columns from the `Orders` table (`o`) will have `NULL` values.
    *   `ON c.CustomerID = o.CustomerID` specifies the join condition, linking orders to their respective customers.

*   **`GROUP BY c.CustomerID, c.CustomerName, c.Email`**:
    *   We group the results by customer details to count orders per unique customer. All non-aggregated columns in the `SELECT` clause must be in the `GROUP BY` clause.

*   **`ORDER BY c.CustomerName`**:
    *   Finally, the results are sorted alphabetically by `CustomerName`.

8.  Retrieve Product Information with Category: Write a SQL query to
display the ProductName, Price, StockQuantity, and CategoryName for all
products. Order the results by CategoryName and then ProductName alphabetically.

````python
print("\n--- Product Information with Category ---")

cursor.execute('''
SELECT
    p.ProductName,
    p.Price,
    p.StockQuantity,
    c.CategoryName
FROM
    Products p
JOIN
    Categories c ON p.CategoryID = c.CategoryID
ORDER BY
    c.CategoryName, p.ProductName;
''')

product_category_results = cursor.fetchall()

# Print header
print(f"{"Product Name":<25} {"Price":<10} {"Stock":<10} {"Category Name":<20}")
print("-" * 65)

# Print results
for row in product_category_results:
    print(f"{row[0]:<25} {row[1]:<10} {row[2]:<10} {row[3]:<20}")
````

### Explanation of the SQL Query:

*   **`SELECT p.ProductName, p.Price, p.StockQuantity, c.CategoryName`**:
    *   We select the `ProductName`, `Price`, and `StockQuantity` from the `Products` table (aliased as `p`).
    *   We select `CategoryName` from the `Categories` table (aliased as `c`).

*   **`FROM Products p JOIN Categories c ON p.CategoryID = c.CategoryID`**:
    *   We perform an `INNER JOIN` between the `Products` table and the `Categories` table.
    *   The join condition `p.CategoryID = c.CategoryID` links each product to its corresponding category based on the `CategoryID`.

*   **`ORDER BY c.CategoryName, p.ProductName`**:
    *   The results are first sorted by `CategoryName` in ascending order.
    *   Then, for products within the same category, they are further sorted by `ProductName` in ascending order.

9.  Write a SQL query that uses a Common Table Expression (CTE) and a
Window Function (specifically ROW_NUMBER() or RANK()) to display the
CategoryName, ProductName, and Price for the top 2 most expensive products in
each CategoryName.

````python
print("\n--- Top 2 Most Expensive Products per Category ---")

cursor.execute('''
WITH RankedProducts AS (
    SELECT
        c.CategoryName,
        p.ProductName,
        p.Price,
        RANK() OVER (PARTITION BY c.CategoryName ORDER BY p.Price DESC) as RankNum
    FROM
        Products p
    JOIN
        Categories c ON p.CategoryID = c.CategoryID
)
SELECT
    CategoryName,
    ProductName,
    Price
FROM
    RankedProducts
WHERE
    RankNum <= 2
ORDER BY
    CategoryName, Price DESC;
''')

top_products_results = cursor.fetchall()

# Print header
print(f"{"Category Name":<20} {"Product Name":<25} {"Price":<10}")
print("-" * 55)

# Print results
for row in top_products_results:
    print(f"{row[0]:<20} {row[1]:<25} {row[2]:<10}")
````

### Explanation of the SQL Query:

*   **`WITH RankedProducts AS (...)`**: This defines a Common Table Expression (CTE) named `RankedProducts`.

*   **Inside the CTE**:
    *   `SELECT c.CategoryName, p.ProductName, p.Price, RANK() OVER (PARTITION BY c.CategoryName ORDER BY p.Price DESC) as RankNum`: We select the `CategoryName`, `ProductName`, and `Price`.
    *   `RANK() OVER (PARTITION BY c.CategoryName ORDER BY p.Price DESC)`: This is the window function. It partitions the data by `CategoryName` (meaning it groups products by their category) and then orders them by `Price` in descending order within each category. `RANK()` assigns a rank to each product, with the most expensive product getting rank 1. If two products have the same price, they will receive the same rank, and the next rank will be skipped (e.g., 1, 1, 3).

*   **Main Query**:
    *   `SELECT CategoryName, ProductName, Price FROM RankedProducts WHERE RankNum <= 2`: We select the relevant columns from our `RankedProducts` CTE and filter the results to only include products whose `RankNum` is 1 or 2 (i.e., the top 2 most expensive products in each category).
    *   `ORDER BY CategoryName, Price DESC`: The final results are ordered by `CategoryName` and then by `Price` in descending order, making it easy to review the top products within each category.
